```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import zipfile
from google.colab import files
import io

!unzip -q DATA_KMEANS_ALG.zip
```

```python

# Load the dataset from the extracted files
df = pd.read_csv('/content/Mall_Customers.csv')



# Display the first few rows to verify the data
df.head()
```

```python
# Drop any null values (optional, depending on your dataset)
df = df.dropna()

# Select relevant features for clustering (modify as needed)
features = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Plotting a pairplot
sns.pairplot(df[['Annual Income (k$)', 'Spending Score (1-100)']], diag_kind='kde')
plt.show()
```

```python
# Apply K-Means with a specified number of clusters (e.g., 5)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_features)

# Add the cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_

# Display the dataframe with the cluster labels
df.head()
```

```python
# Reduce the dimensionality for visualization
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_features[:, 0], y=pca_features[:, 1], hue=df['Cluster'], palette='Set1')
plt.title('Customer Clusters')
plt.show()
```
